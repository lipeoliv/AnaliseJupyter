### Questão 7:
Imprima o nome de todos os pokemons que estão acima ou abaixo do desvio padrão.

```python
data = np.loadtxt('Pokemon.csv', delimiter=',', skiprows=1, usecols=6)
desvio_padrao = np.std(data)

limite_superior = np.mean(data) + desvio_padrao
limite_inferior = np.mean(data) - desvio_padrao

with open('Pokemon.csv', 'r') as arquivo:
    next(arquivo)
    for linha in arquivo:
        dados = linha.strip().split(',')
        nome_pokemon = dados[1]
        ataque = int(dados[6])

        if ataque > limite_superior or ataque < limite_inferior:
            print(nome_pokemon)


            

            
```
