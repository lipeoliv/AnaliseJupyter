## plot()

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
x =[1,2,3,4,5] 
y = [-2,5,200,50,-24]
plt.plot(x,y)
plt.show()
```

### Legendas

```python
x =[1,2,3,4,5] 
y = [-2,5,200,50,-24]
plt.plot(x,y, label="Notas")
plt.legend()
plt.show()
```

```python
x =[1,2,3,4,5] 
y = [-2,5,200,50,-24]
plt.plot(x,y, label="Notas")
plt.legend()
plt.xlabel("Nomde do Eixo x")
plt.ylabel("Nome do Eixo y")
plt.show()
```

### Escala

```python
x =[1,2,3,4,5] 
y = [-2,5,200,50,-24]
plt.plot(x,y, label="Notas")
plt.legend()
plt.xlabel("Nomde do Eixo x")
plt.ylabel("Nome do Eixo y")
plt.xticks([1,2,5])
plt.yticks([-30,0,30,60,90,120,150,180,210])
plt.show()
```

### Propriedades de Linha

https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html

```python
x =[1,2,3,4,5] 
y = [-2,5,200,50,-24]
plt.plot(x,y, label="Notas",linestyle="dashed", color="g")
plt.legend()
plt.xlabel("Nomde do Eixo x")
plt.ylabel("Nome do Eixo y")
plt.xticks([1,2,5])
plt.yticks([-30,0,30,60,90,120,150,180,210])
plt.show()
```

```python
x =[1,2,3,4,5] 
y = [-2,5,200,50,-24]
plt.plot(x,y, label="Notas",linestyle="dashed", color="g",linewidth="3")
plt.legend()
plt.xlabel("Nomde do Eixo x")
plt.ylabel("Nome do Eixo y")
plt.xticks([1,2,5,7])
plt.yticks([-30,0,30,60,90,120,150,180,210])
plt.title("Gráfico da Aula")
plt.show()
```

## Tipos de Gráficos 

### scatter - gráfico de dispersão

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

```python
plt.scatter(x,y)
plt.show()
```

### bar - gráfico de barras

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

```python
plt.bar(x,y)
plt.show()
```

### pie - gráfico de pizza

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html

```python
z = [23,5,100,2]
plt.pie(z)
plt.show()
```

## Mudando o tipo de marcador do gráfico

https://matplotlib.org/stable/api/markers_api.html

```python
plt.scatter(x,y, marker="v")
plt.show()
```

## Mudando os limites do gráfico

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axis.html

```python
plt.plot(x,y)
plt.axis(xmin=-1,xmax=8,ymin=-50,ymax=300)
plt.show()
```

```python
plt.plot(x,y)
plt.axis("square")
plt.show()
```
